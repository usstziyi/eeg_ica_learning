这段话的关键是：

[
\text{白化之后，数据已经被去相关，并且每个方向方差相同}
]

所以 ICA 后面不用再负责“缩放”和“拉伸”，只需要负责“旋转方向”。

---

## 1. 白化前：解混矩阵可以是任意矩阵

假设原始混合信号是：

[
X = AS
]

其中：

[
X
]

是观测到的混合信号，

[
S
]

是源信号，

[
A
]

是混合矩阵。

ICA 想找一个解混矩阵：

[
W
]

使得：

[
S \approx WX
]

如果没有白化，矩阵 (W) 可以是一个一般矩阵。

例如在 3 维情况下：

[
W=
\begin{bmatrix}
w_{11} & w_{12} & w_{13} \
w_{21} & w_{22} & w_{23} \
w_{31} & w_{32} & w_{33}
\end{bmatrix}
]

它有 9 个参数。

这 9 个参数里面，既包含：

[
\text{旋转}
]

也包含：

[
\text{缩放}
]

还包含：

[
\text{剪切}
]

所以搜索空间很大。

---

## 2. 白化后：数据已经被变成“球状”

白化后的数据记作：

[
Z
]

白化的结果是：

[
\operatorname{Cov}(Z)=I
]

也就是：

[
\operatorname{Cov}(Z)
=====================

\begin{bmatrix}
1 & 0 \
0 & 1
\end{bmatrix}
]

或者在 3 维中：

[
\operatorname{Cov}(Z)
=====================

\begin{bmatrix}
1 & 0 & 0 \
0 & 1 & 0 \
0 & 0 & 1
\end{bmatrix}
]

这表示：

[
\text{每个方向方差都是 1}
]

并且：

[
\text{不同方向之间协方差都是 0}
]

所以点云变成了球形，或者二维里是圆形。

---

## 3. 为什么球状分布只允许“旋转”？

因为如果你对白化后的数据再做一个变换：

[
Y = WZ
]

如果希望变换后的数据仍然保持白化性质，也就是：

[
\operatorname{Cov}(Y)=I
]

那么：

[
\operatorname{Cov}(Y)
=====================

\operatorname{Cov}(WZ)
]

因为 (W) 是线性变换，所以：

[
\operatorname{Cov}(Y)
=====================

W\operatorname{Cov}(Z)W^T
]

又因为：

[
\operatorname{Cov}(Z)=I
]

所以：

[
\operatorname{Cov}(Y)
=====================

WIW^T
]

即：

[
\operatorname{Cov}(Y)
=====================

WW^T
]

如果还要求：

[
\operatorname{Cov}(Y)=I
]

那就必须有：

[
WW^T=I
]

这正是正交矩阵的条件。

所以白化后，后续的解混矩阵 (W) 不能随便选，而必须满足：

[
WW^T=I
]

也就是说：

[
W \text{ 必须是正交矩阵}
]

---

## 4. 正交矩阵为什么可以理解为旋转矩阵？

正交矩阵的核心性质是：

[
WW^T=I
]

它不会改变向量长度，也不会改变向量之间的夹角。

也就是说，如果有一个向量：

[
z
]

经过正交矩阵变换：

[
y = Wz
]

那么长度保持不变：

[
|y|=|z|
]

两个向量之间的夹角也保持不变。

这种变换在几何上就是：

[
\text{旋转}
]

或者：

[
\text{旋转 + 镜像翻转}
]

如果进一步要求：

[
\det(W)=1
]

那么它就是纯旋转。

如果：

[
\det(W)=-1
]

则包含镜像翻转。

在 ICA 里，我们通常把它直观理解为“只需要找旋转方向”。

---

## 5. 为什么二维时只需要一个旋转角度？

二维正交旋转矩阵可以写成：

[
R(\theta)
=========

\begin{bmatrix}
\cos\theta & -\sin\theta \
\sin\theta & \cos\theta
\end{bmatrix}
]

你看，这个矩阵虽然有 4 个元素：

[
\cos\theta,\ -\sin\theta,\ \sin\theta,\ \cos\theta
]

但是它们都由同一个参数决定：

[
\theta
]

所以在二维白化数据中，ICA 后续只需要寻找一个角度：

[
\theta
]

也就是找：

[
\text{把圆形点云旋转到哪个角度后，各个投影最独立}
]

这就是“只需要搜索一个旋转角度”的意思。

---

## 6. 但是注意：3 维不是一个角度

你原句里说：

> 原本需要搜索 9 个参数的一般矩阵，现在只需搜索一个旋转角度。

这个说法需要修正。

如果是 3 维，一般矩阵确实有 9 个参数：

[
3 \times 3 = 9
]

但 3 维正交矩阵不是只有 1 个自由度，而是有 3 个自由度。

也就是说，在 3 维中，白化后搜索空间从：

[
9 \text{ 个参数的一般矩阵}
]

降到：

[
3 \text{ 个旋转自由度}
]

而不是 1 个。

“只需要一个旋转角度”只适用于二维情况。

二维中：

[
2 \times 2 \text{ 一般矩阵有 4 个参数}
]

白化后变成：

[
1 \text{ 个旋转角度}
]

三维中：

[
3 \times 3 \text{ 一般矩阵有 9 个参数}
]

白化后变成：

[
3 \text{ 个旋转自由度}
]

更一般地，(n) 维正交矩阵的自由度是：

[
\frac{n(n-1)}{2}
]

所以：

[
n=2 \Rightarrow \frac{2(2-1)}{2}=1
]

[
n=3 \Rightarrow \frac{3(3-1)}{2}=3
]

---

## 7. 放回 FastICA 里理解

FastICA 的目标不是简单让数据不相关，因为白化已经做了这件事。

FastICA 后面真正做的是：

[
\text{在所有旋转方向里，寻找使投影最非高斯的方向}
]

也就是说：

[
y = w^Tz
]

FastICA 要找一个方向 (w)，使得投影 (y) 的非高斯性最大。

因为白化后数据是球状的，所以 (w) 只需要在单位圆或单位球面上转动。

二维时，(w) 可以写成：

[
w=
\begin{bmatrix}
\cos\theta \
\sin\theta
\end{bmatrix}
]

所以寻找 (w)，就等价于寻找一个角度：

[
\theta
]

这就是为什么二维图里可以说：

[
\text{最后只需要搜索旋转角度}
]

---

## 8. 最直观的一句话

白化前，ICA 要同时解决：

[
\text{去相关 + 缩放 + 旋转 + 找独立方向}
]

白化后，前两件事已经完成：

[
\text{去相关 + 单位方差}
]

所以 FastICA 主要只剩下：

[
\text{在球状数据上旋转坐标轴，找到最非高斯的投影方向}
]

因此在二维情况下，问题就从搜索一个任意矩阵，简化成搜索一个旋转角度。
