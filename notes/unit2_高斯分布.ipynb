这里说“更高斯”时，通常不是在看**时间波形图的 y 轴**，而是在看**信号取值的概率分布**。

如果写：

$$
X = AS
$$

其中：

$$
S=
\begin{bmatrix}
s_1(t)\
s_2(t)\
\vdots\
s_n(t)
\end{bmatrix}
$$

$$
X=
\begin{bmatrix}
x_1(t)\
x_2(t)\
\vdots\
x_m(t)
\end{bmatrix}
$$

那么第 $$i$$ 个观测混合信号是：

$$
x_i(t)=a_{i1}s_1(t)+a_{i2}s_2(t)+\cdots+a_{in}s_n(t)
$$

如果画的是**混合信号的时间波形**：

$$
x_i(t)
$$

那么：

* x 轴：时间 $$t$$ 或采样点 $$n$$
* y 轴：混合后的信号幅值/电压/振幅

例如 EEG 中，y 轴通常是：

$$
\mu V
$$

也就是微伏。

但如果你在说 ICA 里的“混合后更高斯”，通常画的是**直方图 / 概率密度图**：

$$
p(x_i)
$$

这时：

* x 轴：信号幅值，也就是 $$x_i$$ 的取值
* y 轴：这个幅值出现的频率、概率密度，或者归一化后的概率

所以此时的 y 轴不是信号幅值，而是：

$$
p(x_i)
$$

也就是“某个幅值出现的概率密度”。

可以这样理解：

时间波形图：

$$
t \rightarrow x_i(t)
$$

看的是“信号随时间怎么变”。

概率分布图：

$$
x_i \rightarrow p(x_i)
$$

看的是“这个信号的取值分布长什么样”。

ICA 说“越高斯”主要是在概率分布意义上说的：混合信号 $$x_i$$ 是多个独立源的加权和，根据中心极限定理，它的取值分布往往比单个非高斯源更接近高斯分布。

所以你这句话里的“更高斯”，对应的 y 轴通常是：

$$
p(x)
$$

即概率密度，而不是 EEG 波形里的电压幅值。
