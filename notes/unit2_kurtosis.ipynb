## 峰度（Kurtosis）的含义

`kurtosis(data)` 来自 `scipy.stats`（第 54 行），用于计算数据分布的**峰度（Kurtosis）**。

### 峰度的统计学意义

峰度是描述概率分布**尾部厚度**和**尖峰程度**的统计量。`scipy.stats.kurtosis` 默认使用的是**超额峰度（excess kurtosis）**：

| 峰度值 | 分布类型 | 含义 |
|--------|---------|------|
| **= 0** | 高斯分布（正态分布） | 和高斯分布一样的尾部/峰度 |
| **> 0** | 超高斯分布（super-Gaussian） | 比高斯分布更"尖"，尾部更厚（如拉普拉斯分布、EEG信号） |
| **< 0** | 亚高斯分布（sub-Gaussian） | 比高斯分布更"平"（如均匀分布） |

### 在 ICA 中的核心作用

在该 Notebook 的上下文中，峰度是**非高斯性的经典度量**，直接用于解释 FastICA 的核心思想：

1. **中心极限定理**表明：多个独立信号的**混合**（加权和）会比任一单信号**更高斯**（峰度更接近 0）
2. ICA 的思路是反过来：**找到使峰度绝对值最大（最非高斯）的方向 → 那就是源信号的方向**

所以代码中 `kurtosis(data)` 计算的就是数据的超额峰度值，用于量化"离高斯有多远"。数值越偏离 0，说明该信号越可能是分离后的独立源成分。
