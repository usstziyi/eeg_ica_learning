## AU = **Arbitrary Units**（任意单位）

颜色条显示的数值**没有物理意义**，这是 ICA 本身的数学性质决定的。

---

### 为什么没有物理单位？

ICA 分解有一个**固有的"符号-幅度模糊性"**：

```
观测数据 X = A × S
              ↑     ↑
          混合矩阵  独立成分

你可以同时：
  A → A × 2（混合权重翻倍）
  S → S ÷ 2（成分信号减半）
  乘积 X 不变！
```

无论 A 和 S 怎么缩放，乘积 X 永远等于原始数据。所以 ICA 的权重（即 colorbar 显示的值）本身没有绝对大小意义——它只是一个相对比例。

---

### 那看什么？

| 别纠结 | 重点看 |
|--------|--------|
| colorbar 上的具体数值（如 0.3 vs 0.5） | **亮斑在哪**（额部？颞部？枕部？） |
| 权重是正还是负 | **分布是否对称**，是否局灶 |
| 不同成分间权重大小对比 | **是否符合已知伪迹的空间特征** |

---

### 举例

```
成分 0：
  前额（EEG 001/002 附近）大亮斑 → 眨眼伪迹 ✅
  colorbar 显示 "0.4 AU" → 不用管，没意义

成分 3：
  双侧颞部亮斑 → 肌电伪迹 ✅
  colorbar 显示 "-0.3 AU" → 不用管
  
成分 7：
  全头皮均匀分布 → 可能是心跳 or 真实脑电
  需要结合时间序列和 PSD 进一步判断
```

---

### 一句话总结

> **AU 就是"无所谓单位"的意思。ICA topomap 的 colorbar 数值没有绝对含义，你只需要关注亮斑的位置和形状——那才是判断伪迹的关键。**

## ICA 地形图配色解读

ICA 地形图本质上显示的是**混合矩阵 A 中每个成分对应的列向量**，即该成分"投射到 60 个通道上的权重值"。

---

### 颜色代表什么？

一句话：**该成分在某个电极位置上被"贡献"了多少信号，以及极性方向。**

| 颜色 | 含义 |
|------|------|
| **红色 / 暖色** | **正值**：该成分的电信号在该位置是正向贡献（相对于平均参考） |
| **蓝色 / 冷色** | **负值**：该成分的电信号在该位置是负向贡献 |
| **白色 / 中间色** | **≈ 0**：该成分在这个位置几乎没贡献 |
| **深色（深红 / 深蓝）** | **权重大**：该成分在这个通道上影响强 |
| **浅色（浅红 / 浅蓝）** | **权重小**：影响弱 |

---

### 为什么会有红有蓝？（极性）

EEG 是差分信号（相对于参考电极），物理上脑电信号在头皮上天然就是"一边正，一边负"的偶极子分布：

```
        正电荷堆积                → 红色
    ┌─────⊕─────┐
    │           │
    │   电流流   │               ← 偶极子
    │   动方向  │
    │           │
    └─────⊖─────┘
        负电荷堆积                → 蓝色
```

所以一个成分在 topomap 上**出现红蓝对称亮斑**是完全正常的——那是一个偶极子，不是伪迹。反过来，如果只有一片孤立的红（没有对应的蓝），那通常不是真实脑电来源，更像伪迹。

---

### 各类伪迹在 ICA 地形图上的典型分布

| 伪迹类型 | Topomap 特征 |
|----------|-------------|
| **眨眼** | 前额（Fp1/Fp2 附近）**孤立大红/蓝亮斑**，没有对称偶极子，颜色很深 |
| **水平眼动** | 额头两侧（F7/F8）**一红一蓝**，几乎对称的偶极子对 |
| **心跳** | 全头皮**均匀分布**，颜色较浅，无局灶特征 |
| **肌电（咬牙）** | 双侧颞部（耳朵上方）**局灶性深色亮斑** |
| **肌电（颈部紧张）** | 枕部 / 后颞部（O1/O2/T5/T6）局灶亮斑 |
| **真实脑电（如 α 节律）** | 枕部（O1/O2）**红蓝对称偶极子**，深浅适中 |

---

### 判断口诀

```
额头一抹红 → 眨眼
额头左右红蓝对称 → 横眼
耳边颜色深 → 咬牙/肌电
全头均匀洒 → 心跳
枕部对称偶极子 → 可能是真实 α 节律
```
