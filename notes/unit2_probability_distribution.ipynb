### 均匀分布

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-0.5, 1.5, 500)
y = np.where((x >= 0) & (x <= 1), 1.0, 0.0)

fig, ax = plt.subplots(figsize=(8, 4))
ax.fill_between(x, y, alpha=0.3, color="#3b82f6")
ax.plot(x, y, color="#3b82f6", linewidth=2)
ax.set_title("Uniform Distribution  U(0, 1)", fontsize=14, fontweight="bold")
ax.set_xlabel("x")
ax.set_ylabel("Probability Density")
ax.set_ylim(-0.1, 1.5)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("uniform.png", dpi=150)
plt.show()
```

### 拉普拉斯分布

```python
import numpy as np
import matplotlib.pyplot as plt

mu, b = 0, 1
x = np.linspace(-6, 6, 500)
y = (1 / (2 * b)) * np.exp(-np.abs(x - mu) / b)

fig, ax = plt.subplots(figsize=(8, 4))
ax.fill_between(x, y, alpha=0.3, color="#ef4444")
ax.plot(x, y, color="#ef4444", linewidth=2)
ax.set_title("Laplace Distribution  (mu=0, b=1)", fontsize=14, fontweight="bold")
ax.set_xlabel("x")
ax.set_ylabel("Probability Density")
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("laplace.png", dpi=150)
plt.show()
```

### 高斯分布

```python
import numpy as np
import matplotlib.pyplot as plt

mu, sigma = 0, 1
x = np.linspace(-5, 5, 500)
y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

fig, ax = plt.subplots(figsize=(8, 4))
ax.fill_between(x, y, alpha=0.3, color="#10b981")
ax.plot(x, y, color="#10b981", linewidth=2)
ax.set_title("Gaussian Distribution  (mu=0, sigma=1)", fontsize=14, fontweight="bold")
ax.set_xlabel("x")
ax.set_ylabel("Probability Density")
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("gaussian.png", dpi=150)
plt.show()
```

## 三种分布的数学表达式

### 1. 均匀分布 (Uniform Distribution)

$$f(x) = \begin{cases} \dfrac{1}{b - a}, & a \leq x \leq b \\[6pt] 0, & \text{otherwise} \end{cases}$$

记作 $X \sim U(a, b)$

---

### 2. 拉普拉斯分布 (Laplace Distribution)

$$f(x) = \frac{1}{2b} \exp\!\left(-\frac{|x - \mu|}{b}\right)$$

记作 $X \sim \text{Laplace}(\mu, b)$

- $\mu$：位置参数（均值）
- $b > 0$：尺度参数（控制分散程度）

---

### 3. 高斯分布 (Gaussian / Normal Distribution)

$$f(x) = \frac{1}{\sqrt{2\pi}\,\sigma} \exp\!\left(-\frac{(x - \mu)^2}{2\sigma^2}\right)$$

记作 $X \sim \mathcal{N}(\mu, \sigma^2)$

- $\mu$：均值
- $\sigma > 0$：标准差

---

### 对比记忆

| | 核心衰减方式 | 关键区别 |
|---|---|---|
| **Uniform** | 无衰减，常数 | 有限支撑，区间外为 0 |
| **Laplace** | $\exp(-\|x\|)$ 指数衰减 | 绝对值 → 尖峰、厚尾 |
| **Gaussian** | $\exp(-x^2)$ 二次指数衰减 | 平方 → 圆润钟形、薄尾 |

Laplace 的 $|x-\mu|$ vs Gaussian 的 $(x-\mu)^2$ 是两者最本质的区别——前者导致中点处不可导（尖角），后者处处光滑。
